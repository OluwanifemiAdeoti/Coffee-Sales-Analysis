```python
coffee_sales["coffee_name"] = coffee_sales["coffee_name"].astype("category")
```

```python
coffee_sales["payment_method"] = coffee_sales["payment_method"].astype("category")
```

```python
coffee_sales["hour"]=coffee_sales["datetime"].dt.hour
```

```python
coffee_sales["day"]=coffee_sales["datetime"].dt.day_name()
```

```python
def time_of_day(hour):
    if hour < 12:
        return "Morning"
    elif hour < 17:
        return "Afternoon"
    else:
        return "Evening"
```

```python
coffee_sales["time_of_day"] = coffee_sales["hour"].apply(time_of_day)
```
